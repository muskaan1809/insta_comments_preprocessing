```
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
text = "Hello everyone! learned nlp from sandip sir"
```

```
tokens = nltk.word_tokenize(text)
print(tokens)
tag = nltk.pos_tag(tokens)
print(tag)
grammar = "NP: {<DT>?<JJ>*<NN>}"
cp  =nltk.RegexpParser(grammar)
result = cp.parse(tag)
print(result)
```

```
import spacy
sp = spacy.load('en_core_web_sm')

#To see the dependency of each token on another along with the POS tag.
from spacy import displacy
sen = sp(text)
displacy.render(sen, style='dep', jupyter=True, options={'distance': 85})
```
